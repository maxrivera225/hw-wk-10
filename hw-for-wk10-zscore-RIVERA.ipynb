```python
import pandas as pd
from IPython.display import display, HTML
## run the display code here
pd.options.display.float_format = '{:,.3f}'.format

## small trick to improve our display
## will allow us to see dataframes side-by-side
from IPython.display import display, HTML
```

```python
dfs = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv")
dfs

lps = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv")
lps
```

```python
dfs
```

```python
from scipy.stats import zscore
```

```python
dfs["fs_zscore"] = zscore(dfs["streams_2023"])
```

```python
dfs
dfs["mod_streams"] = dfs.apply((lambda x: x["streams_2023"] / 1500), axis = 1)
```

```python
dfs["mod_zscore"] = zscore(dfs["mod_streams"])
```

```python
dfs
```

```python
lps["lp_zscore"] = zscore(lps["sales"])
```

```python
lps.head(10)
```

```python
from sklearn.preprocessing import normalize

## pd.options.display.float_format = '{:,.3f}'.format
new_df = pd.concat([lps, dfs])
```

```python
new_df
```

```python
new_df.to_csv("new_df_1.csv")
```

```python
## The album that stood out the most was Sgt. Peppers Lonely Hearts Club by The Beatles - almost 8 above the std
```

```python
df = pd.read_csv("new_df_2.csv")
```

```python
top_ten = df.sort_values("sales", ascending = False).head(10)
```

```python
top_ten["zscore"] = zscore(top_ten["sales"])
```
